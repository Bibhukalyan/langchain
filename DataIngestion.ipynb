#### 4) a) FAISS

```python
from langchain_community.vectorstores import FAISS
from langchain_community.document_loaders import TextLoader
from langchain_huggingface import HuggingFaceEmbeddings #OllamaEmbeddings
from langchain_text_splitters import CharacterTextSplitter

loader = TextLoader('speech.txt')
documents = loader.load()

text_splitter = CharacterTextSplitter(chunk_size=200, chunk_overlap=30)
docs = text_splitter.split_documents(documents)
```

```python
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
db = FAISS.from_documents(docs, embeddings)
db
```

```python
### querying
query = "What does the speaker believe is the main reason the United States should enter the war?"
# query = "How does the speaker describe the desired outcome of the war?"
docs = db.similarity_search(query)
docs
```

```python
docs[0].page_content
```

#### As a Retriver

```python
retriver = db.as_retriever()
docs=retriver.invoke(query)
```

```python
docs[0].page_content
```

## Similiarity search with score

#### Lowe the score is better

```python
docs_and_scores = db.similarity_search_with_score(query)
docs_and_scores
```

```python
#Embeding search
embedding_search = embeddings.embed_query(query)
embedding_search
```

```python
docs_score=db.similarity_search_by_vector(embedding_search)
docs_score
```

```python
### saving and loading
db.save_local("faiss_index")
```

```python
new_db = FAISS.load_local("faiss_index", embeddings, allow_dangerous_deserialization=True)
new_db
```

```python
docs=new_db.similarity_search(query)
docs
```

## Chroma DB

```python
#building a sample vectordb
from langchain_community.vectorstores import Chroma
from langchain_community.document_loaders import TextLoader
from langchain_huggingface import HuggingFaceEmbeddings #OllamaEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter

loader = TextLoader('speech.txt')
documents = loader.load()

text_splitter = CharacterTextSplitter(chunk_size=500, chunk_overlap=30)
docs = text_splitter.split_documents(documents)

embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
db = Chroma.from_documents(documents=docs, embedding=embeddings, persist_directory="./chroma_db") #to save in local storage DB
db
```

```python
### querying
query = "What does the speaker believe is the main reason the United States should enter the war?"
# query = "How does the speaker describe the desired outcome of the war?"
docs = db.similarity_search(query)
docs[0].page_content
```

```python
#load DB from persistant directory
db = Chroma(persist_directory="./chroma_db", embedding_function=embeddings)
db
```

```python
### Retriver version
retriver = db.as_retriever()
docs=retriver.invoke(query)
docs[0].page_content
```

## Building simple aplication with LangChain

```python
import os
from dotenv import load_dotenv
load_dotenv() #load all environment variables

os.environ['OPENAI_API_KEY'] = os.getenv('OPENAI_API_KEY')
#Langsmith Tracing
os.environ['LANGCHAIN_API_KEY'] = os.getenv('LANGCHAIN_API_KEY')
os.environ['LANGCHAIN_TRACING_V2'] = "true"
os.environ['LANGCHAIN_PROJECT'] = os.getenv('LANGCHAIN_PROJECT')
```

```python
from langchain_openai import ChatOpenAI

llm=ChatOpenAI()
print(llm)
```

```python
result = llm.invoke("What is the meaning of life?")
result
```

```python
#chatprompt template

from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "You are a helpful assistant."),
        ("user", "{input}")
    ]
)
prompt
```

```python
#chain - 1st it goes to prompt and later result is passed to LLM
chain=prompt | llm
response=chain.invoke({"input":"What is the meaning of life?"})
print(response)
```

```python
# String output parser
from langchain_core.output_parsers import StrOutputParser

chain=prompt | llm | StrOutputParser()
response=chain.invoke({"input":"What is the meaning of life?"})
print(response)
```
